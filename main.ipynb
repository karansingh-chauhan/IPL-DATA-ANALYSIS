```python
train_df['batting_team'].unique()
```

some of the teams are have 2 value so we will convert these into lower case teams

```python
train_df['batting_team'].nunique()
```

```python
batting_dic={
            'Kolkata Knight Riders':'KKR',
            'Royal Challengers Bangalore':'RCB',
            'Chennai Super Kings':'CSK',
            'Kings XI Punjab':'KXP',
            'Rajasthan Royals':'RR',
            'Delhi Daredevils':'DC',
            'Mumbai Indians':'MI',
            'Deccan Chargers':'DEC',
            'Kochi Tuskers Kerala':'KTK',
            'Pune Warriors':"PW",
            'Sunrisers Hyderabad':'SRH',
            'Rising Pune Supergiants':'RPS',
            'Gujarat Lions':'GL',
            'Rising Pune Supergiant':'RPS',
            'Delhi Capitals':'DC',
            'Punjab Kings':'KXP',
            'Lucknow Super Giants':'LSG',
            'Gujarat Titans':'GT',
            'Royal Challengers Bengaluru':'RCB'}
```

```python
# train_df.apply(lambda x : x .replace(batting_dic))
# this process can be slow so insted of using lamda we are going to use the replace function of pandas which is faster or the map function 
# Create a list of columns that contain team names
team_cols = ['batting_team', 'bowling_team'] 

# Option A: Using replace (Great for whole DataFrames or specific columns)
# Strip spaces and replace names for both columns efficiently
for col in team_cols:
    train_df[col] = train_df[col].str.strip().replace(batting_dic)# we added the strip fuction so that the there is no extrea space in
# the daataset which can cause problem in the future
```
